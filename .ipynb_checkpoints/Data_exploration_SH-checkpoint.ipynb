```python
yrs = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
mos = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
wks = ['28']
def bestseller_dates(years, months, weeks):
    d_list = []
    for year in years:
        for month in months:
            for week in weeks:
                bestsellers_date = f'{year}-{month}-{week}'
                d_list.append(bestsellers_date)
    return d_list
```

```python
dates_list = bestseller_dates(yrs, mos, wks)
# dates_list
len(dates_list)
```

```python
yr_2012 = bestseller_dates(yrs[:1], mos, wks)
yr_2013 = bestseller_dates(yrs[1:2], mos, wks)
yr_2014 = bestseller_dates(yrs[2:3], mos, wks)
yr_2015 = bestseller_dates(yrs[3:4], mos, wks)
yr_2016 = bestseller_dates(yrs[4:5], mos, wks)
yr_2017 = bestseller_dates(yrs[5:6], mos, wks)
yr_2018 = bestseller_dates(yrs[6:7], mos, wks)
yr_2019 = bestseller_dates(yrs[7:8], mos, wks)
```

```python
#2012 bestsellers
books_2012 = {}
for date in yr_2012:
    credentials = {'date': date, 'list': 'combined-print-and-e-book-fiction'}
    books_2012.update({date: requests.get("https://api.nytimes.com/svc/books/v3/lists.json?api-key=Z0yWljtxxgI8qc13dN3ZW6gvLM5apn2G", credentials)})
books_2012
```

```python
books = {}
for k, v in books_2012.items():
    books.update({k:v.json()})
```
